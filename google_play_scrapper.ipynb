```python
#import required library
import datetime
import time
import pandas as pd
from tqdm.notebook import tqdm

#data visualization library
import plotly.graph_objects as go
from plotly.subplots import make_subplots

#library for scrapping google play store data
from google_play_scraper import Sort, reviews, reviews_all, app, search
```

```python
#define a list of search terms
search_terms = ['GrabFood', 'ShopeeFood', 'GoFood']

#create empty list and set for unique app id
search_data_list = []
search_app_id_set = set()

#loop over search terms and fetch app data
for search_term in tqdm(search_terms):
    #request and fetch search data
    search_results = search(
        search_term,    #pass down list of search terms
        lang='id',      #default is set to en
        country='id',   #default is set to us
        n_hits=5        #default is set to 30(max allowed)
    )

    #find app id in result
    for search_result in search_results:
        search_app_id = search_result['appId']

        #check for if app_id is unique then append the result and add app_id to search_app_id_set
        if search_app_id not in search_app_id_set:
            search_data_list.append(search_result)
            search_app_id_set.add(search_app_id)
```

```python
#check the data retrieved
search_data_list[0]
```

```python
#convert the data into dataframe and return dataframe with specific column
search_info_df = pd.DataFrame(search_data_list)
search_info_df = search_info_df[['appId', 'icon', 'title', 'score', 'genre', 'installs', 'developer']]

#just personal preference to rename the column name for appId and installs
search_info_df = search_info_df.rename(columns={'appId': 'app_id', 'installs': 'install'})

#check the data head
search_info_df.head()
```

```python
#modify the column type for install and score
search_info_df['score'] = search_info_df['score'].apply(lambda x: round(float(x), 2))
search_info_df['install'] = search_info_df['install'].apply(lambda x: int(x.replace('.', '').replace('+', '')))

#check the dataframe
search_info_df.info()
```

```python
#check the result
search_info_df.head()
```

```python
#define function for plotting data search and data app
def search_app_plot(df, plot_title):
    #create subplots and add 2 trace for rating and install
    fig = make_subplots(specs=[[{'secondary_y': True}]])
    fig.add_trace(go.Bar(x=df['title'], y=df['score'], name='Rating'), secondary_y=False)
    fig.add_trace(go.Scatter(x=df['title'], y=df['install'], name='Install'), secondary_y=True)

    #update layout with legend
    fig.update_layout(title=plot_title,
                      xaxis_title='Apps',
                      yaxis_title='Rating',
                      yaxis2_title='install',
                      legend=dict(x=0.5, y=1.25,
                                  xanchor='center', orientation='h',
                                  traceorder='normal',
                                  font=dict(size=10))
                      )
    fig.show()
```

```python
#call function to plot data search
search_app_plot(search_info_df, 'Rating and Install in Search Result')
```

```python
#export data into csv
search_info_df.to_csv('search_info_data.csv', index=False)
```

```python
#define a list of app id
app_ids = ['com.grabtaxi.passenger', 'com.shopee.id', 'com.gojek.app']

#create empty list
app_data_list = []

#loop over app id and fetch app data
for app_id in tqdm(app_ids):
    #request and fetch app data
    app_result = app(
        app_id,         #pass down list of app id
        lang='id',      #default is set to en
        country='id'    #default is set to us
    )

    #append the result
    app_data_list.append(app_result)
```

```python
#create empty dictionary to store each app id review data
review_data_dict = {}

#iterate through app_ids list
for app_id in tqdm(app_ids):
    try:
        #initialize review data list, continuation token, review count, non match review count
        review_data = []
        continuation_token = None
        matching_review_count = 0
        non_matching_review_count = 0

        #set the date range for review data retrieval (format=Y, M, D, H, M, S)
        start_date = datetime.datetime(2024, 1, 1, 0, 0, 0)
        end_date = datetime.datetime(2024, 5, 5, 23, 59, 59)

        #initialize tqdm progress bar for non-matching review count
        review_bar = tqdm(desc=f"Review count for {app_id}: ", total=100)

        #fetch review data
        while True:
            #fetch review data with continuation token
            review_result, continuation_token = reviews(
                app_id,                                 #pass down app id
                lang='id',                              #default="en"
                country='id',                           #default='us'
                sort=Sort.NEWEST,                       #default=Sort.NEWEST
                count=100,                              #default=100 (max=200)
                continuation_token=continuation_token   #pass down the continuation token
            )

            #filter the reviews based on the date range and extend the data
            filtered_reviews = [review for review in review_result if start_date <= review['at'] <= end_date]
            review_data.extend(filtered_reviews)

            #check if any review pass/in certain range
            for review in review_result:
                if review['at'] < start_date:
                    non_matching_review_count += 1
                elif start_date <= review['at'] <= end_date:
                    matching_review_count += 1
                    non_matching_review_count = 0
                else:
                    non_matching_review_count = 0

            #update tqdm progress bar for review count
            review_bar.update(1)
            review_bar.set_postfix({"Matched Reviews": matching_review_count,
                                    "Non-matched Reviews": non_matching_review_count})

            #if there is no continuation token left or no matching results reach 50000, break the loop
            if not continuation_token or non_matching_review_count >= 500:
                break

        #close tqdm progress bar for non-matching review count
        review_bar.close()

        #store the review data in a dataframe and add it to the dictionary with the app_id as key
        review_data_dict[app_id] = pd.DataFrame(review_data)

        #added time delay before another app review request
        time.sleep(5)

    #error handling
    except Exception as e:
        print(f'Error fetching reviews for app ID {app_id}: {e}')
```

```python
#display the first few rows of each dataframe and check the result
for app_id, df in review_data_dict.items():
    print(f'{app_id}:')
    display(df.head(1))
```

```python
#iterate over the dictionary
for app_id, df in review_data_dict.items():
    #rename the columns of the dataframe
    df = df.rename(columns={'reviewId': 'review_id',
                            'userName': 'username',
                            'userImage': 'user_icon',
                            'thumbsUpCount': 'helpfulness',
                            'reviewCreatedVersion': 'created_version',
                            'at': 'date',
                            'replyContent': 'reply',
                            'repliedAt': 'reply_date',
                            'appVersion': 'app_version'}
                            )

    #update the dataframe inside the dictionary
    review_data_dict[app_id] = df

#try accessing the data with keys(app_id)
for app_id, df in review_data_dict.items():
    print(f'{app_id}: {df.shape[0]} reviews')
```

```python
#select dataframe by 'date' and calculate the mean 'score'
#for grab
review_data_grab = review_data_dict['com.grabtaxi.passenger'].groupby(pd.Grouper(key='date', freq='d'))['score'].mean().reset_index()
review_data_grab['date'] = review_data_grab['date'].dt.strftime('%d')

#for shopee
review_data_shopee = review_data_dict['com.shopee.id'].groupby(pd.Grouper(key='date', freq='d'))['score'].mean().reset_index()
review_data_shopee['date'] = review_data_shopee['date'].dt.strftime('%d')

#for gojek
review_data_gojek = review_data_dict['com.gojek.app'].groupby(pd.Grouper(key='date', freq='d'))['score'].mean().reset_index()
review_data_gojek['date'] = review_data_gojek['date'].dt.strftime('%d')
```

```python
#define traces for each app
trace1 = go.Scatter(x=review_data_shopee['date'], y=review_data_shopee['score'], name='Shopee')
trace2 = go.Scatter(x=review_data_gojek['date'], y=review_data_gojek['score'], name='Gojek')
trace3 = go.Scatter(x=review_data_grab['date'], y=review_data_grab['score'], name='Grab')

#create a subplot figure with one row and one column
fig = make_subplots(rows=1, cols=1)

#add the traces to the figure
fig.add_trace(trace1)
fig.add_trace(trace2)
fig.add_trace(trace3)

#update the figure layout with title and axis labels
fig.update_layout(title='App Review Ratings Over Time',
                  xaxis_title='Date',
                  yaxis_title='Rating')
fig.show()
```

```python
#export data from dictionary as csv
for app_id, df in review_data_dict.items():
    print(f'Exporting review data {app_id} to csv')
    df.to_csv(f'review_data_{app_id}.csv', index=False)
```
